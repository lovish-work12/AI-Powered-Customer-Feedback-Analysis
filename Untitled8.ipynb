```python
import pandas as pd
from transformers import pipeline
```

```python
df = pd.read_csv("feedback.csv")
print("Sample Feedback:\n", df.head())
```

```python
sentiment_model = pipeline("sentiment-analysis")
```

```python
df["Sentiment"] = df["Customer Feedback"].apply(lambda x: sentiment_model(x)[0]['label'])
```

```python
df.to_csv("feedback_results.csv", index=False)
```
